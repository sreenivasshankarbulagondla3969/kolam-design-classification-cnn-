**🔹 Step 2: Scraping Script**

```python
from icrawler.builtin import GoogleImageCrawler
import os

# Define kolam types (classes)
kolam_types = {
    "Pulli Kolam": "pulli kolam design",
    "Chikku Kolam": "chikku kolam pattern",
    "Sikku Kolam": "sikku kolam drawing",
    "Rangoli Kolam": "rangoli kolam color",
    "Freehand Kolam": "freehand kolam design",
    "Mandala Kolam": "mandala kolam"
}

# Number of images per type
num_images = 100

for folder, query in kolam_types.items():
    # Ensure the root_dir exists before initializing the crawler
    storage_path = f"dataset/{folder}"
    os.makedirs(storage_path, exist_ok=True)

    crawler = GoogleImageCrawler(
        storage={"root_dir": storage_path},
        downloader_threads=64,
        parser_threads=64,
    )
    crawler.crawl(keyword=query, max_num=num_images)
```

**🔹 Step 4: Clean the Dataset (Optional)**

```python
import os
from PIL import Image

def clean_dataset(path):
    if not os.path.exists(path):
        print(f"Directory '{path}' not found. Skipping cleaning.")
        return
    for folder in os.listdir(path):
        folder_path = os.path.join(path, folder)
        for img_name in os.listdir(folder_path):
            try:
                img_path = os.path.join(folder_path, img_name)
                img = Image.open(img_path)
                if img.size[0] < 50 or img.size[1] < 50:  # remove tiny images
                    os.remove(img_path)
            except:
                os.remove(img_path)  # remove corrupt files

clean_dataset("dataset")
```
