**Kolam Design Identification using deep learning CNN Algorithm**

**Step 1: Define Kolam Types**

```python
kolam_types = [
    "Pulli Kolam (Dot Kolam)",
    "Chikku/Sikku Kolam",
    "Rangoli (Color Kolam)",
    "Freehand Kolam",
    "Mandala Kolam"
]
```

**Step 2: Basic Code to Train a Classifier**

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# Path where your kolam dataset is stored
# Directory structure should be:
# dataset/
#   Pulli/
#   Chikku/
#   Rangoli/
#   Freehand/
#   Mandala/
data_dir = "dataset/"

# Preprocessing images
img_size = (128, 128)
batch_size = 32

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_gen = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    subset="training",
    class_mode="categorical"
)

val_gen = datagen.flow_from_directory(
    data_dir,
    target_size=img_size,
    batch_size=batch_size,
    subset="validation",
    class_mode="categorical"
)

# Building CNN model
model = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(128,128,3)),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D(2,2),
    layers.Conv2D(128, (3,3), activation="relu"),
    layers.MaxPooling2D(2,2),
    layers.Flatten(),
    layers.Dense(128, activation="relu"),
    layers.Dense(len(train_gen.class_indices), activation="softmax")
])

model.compile(optimizer="adam",
              loss="categorical_crossentropy",
              metrics=["accuracy"])

# Training
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=10
)

# Save model
model.save("kolam_classifier.h5")

# Plot training performance
plt.plot(history.history["accuracy"], label="train acc")
plt.plot(history.history["val_accuracy"], label="val acc")
plt.legend()
plt.show()
```

**🔹 Step 3: Predict Kolam Type from New Image**

```python
import numpy as np
from tensorflow.keras.preprocessing import image
import tensorflow as tf # Import tensorflow here

# Load saved model
model = tf.keras.models.load_model("kolam_classifier.h5")

def predict_kolam(img_path):
    img = image.load_img(img_path, target_size=(128,128))
    img_array = image.img_to_array(img)/255.0
    img_array = np.expand_dims(img_array, axis=0)

    prediction = model.predict(img_array)
    class_idx = np.argmax(prediction)

    # The train_gen variable is not defined in this cell.
    # Assuming the order of classes is the same as defined in kolam_types
    labels = [
        "Pulli Kolam (Dot Kolam)",
        "Chikku/Sikku Kolam",
        "Rangoli (Color Kolam)",
        "Freehand Kolam",
        "Mandala Kolam"
    ]
    return labels[class_idx]

# Example
print(predict_kolam("test_kolam.jpg"))
```
